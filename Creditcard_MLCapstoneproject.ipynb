**INTRODUCTION**:
A bank's credit card department is one of the top adopters of data science. A top focus for the bank has always been acquiring new credit card customers. Giving out credit cards without doing proper research or
evaluating applicants' creditworthiness is quite risky. The credit card department has been using a data-driven system for credit assessment called Credit Scoring for many years, and the model is known as an
application scorecard. A credit card application's cutoff value is determined using the application scorecard, which also aids in estimating the applicant's level of risk. This decision is made based on strategic priority
at a given time.

Customers must fill out a form, either physically or online, to apply for a credit card. The application data is used to evaluate the applicant's creditworthiness. The decision is made using the application data in addition
to the Credit Bureau Score, such as the FICO Score in the US or the CIBIL Score in India, and other internal information on the applicants. Additionally, the banks are rapidly taking a lot of outside data into account to
enhance the caliber of credit judgements.

```python
# Changing the Birthday_count to Age_years
df["Birthday_count"]= abs(round(df["Birthday_count"]/365,0))
# Changing Employed_days to Years_employed
df["Employed_days"]= np.where(df["Employed_days"]>0,0,df["Employed_days"])
df["Employed_days"]= abs(round(df["Employed_days"]/365,0))
# Renaming both Birthday_count and Employed_days
df.rename(columns= {"Birthday_count": "Age_years", "Employed_days": "Years_employed"},inplace=True)
```

The two columns Birthday_count and Employed_days were provided as negative and as days till date thus we did the following:
1. In order for age of a person to be more interpretable we changed it to years, removed negative values and renamed column as Age_years.
2. In employed days columns the negative days meant total number of employed days and if employed days is positive then the person is unemployed so we changed the column name to Years_employed and replaced positive values with zero and removed negative sign from all records.

# Imputation of Null values

```python
df["Gender"]= df["Gender"].fillna(df["Gender"].mode()[0])
df["Annual_income"]= df["Annual_income"].fillna(df["Annual_income"].mode()[0])
df["Age_years"]= df["Age_years"].fillna(df["Age_years"].mean())
df["Type_Occupation"]= df["Type_Occupation"].replace(np.NaN,"Not Specified")
```

1. Imputating Null values in Gender and Annual_income columns with mode of the respective columns.
2. Imputating Null values in Age_years column with mean value of the column.
3. For Type_Occupation there were more than 30 percent null values therefore I replaced null values with "Not Specified"

# Defining a Function for calculation of Metrics

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,roc_auc_score,accuracy_score, log_loss
def modelperformance(y_pred, y_prob):
    accuracy = accuracy_score(y_test, y_pred)
    logloss = log_loss(y_test, y_prob)
    classification_rep = classification_report(y_test, y_pred)
    confusion_mat = confusion_matrix(y_test, y_pred)
    print(f"Accuracy Score: {accuracy}")
    print((f"Log Loss: {logloss}"))
    print(f"Classification Report:\n {classification_rep}")
    print(f"Confusion Matrix:\n {confusion_mat}")
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred, drop_intermediate = False)
    auc = roc_auc_score(y_test, y_pred)

    plt.figure(figsize=(5, 5), dpi=100)
    plt.axis('scaled')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.title("AUC & ROC Curve")
    plt.plot(false_positive_rate, true_positive_rate, 'g')
    plt.fill_between(false_positive_rate, true_positive_rate, facecolor='lightgreen', alpha=0.8)
    plt.text(0.95, 0.05, 'AUC = %0.4f' % auc, ha='right', fontsize=12, weight='bold', color='blue')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()

    # Return performance metrics
    return {"Accuracy Score": accuracy,"Log Loss": logloss, "Classification Report": classification_rep, "Confusion Matrix": confusion_mat, "AUC": auc}
```

```python
conn.execute("SELECT Education, COUNT(*) AS total_count FROM data WHERE (Education= 'Academic degree') GROUP BY Education;").fetchdf()
```
